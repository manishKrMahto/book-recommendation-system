# Book Recommendation System
#### using Collebractive filtering

```python
import numpy as np
import pandas as pd
import pickle

from sklearn.metrics.pairwise import cosine_similarity
```

# loading datasets

```python
books = pd.read_csv('/content/drive/MyDrive/Datasets/book recommendation dataset/Books.csv')
ratings = pd.read_csv('/content/drive/MyDrive/Datasets/book recommendation dataset/Ratings.csv')
```

## knowing data

```python
print(f'book shape : {books.shape}')
print(f'ratings shape : {ratings.shape}')
```

```python
books.head()
```

```python
ratings.head()
```

## joinging or merging books and ratings

```python
new_df = books.merge(ratings , on =['ISBN'])
new_df.head()
```

# Popular based Filtering

Filtering books based on their average ratings, selecting the top 50, and showing them as trending books.

```python
# finding top 50 books that have highest average rating
popular_books = new_df.groupby('Book-Title')['Book-Rating'].mean().reset_index().sort_values('Book-Rating', ascending = False ).head(50)

# renaming 'Book-Rating' to 'Avg-Rating'
popular_books.rename(columns = {'Book-Rating' : 'Avg-Rating'} , inplace =True)
```

```python
popular_books
```

```python
# merging with books to get more columns
top_50_books = popular_books.merge(books , on ='Book-Title').drop_duplicates(subset = 'Book-Title')[['Book-Title' ,'Book-Author' , 'Image-URL-M' , 'Avg-Rating']]
```

```python
top_50_books
```

```python
# extracting the file to build web page
pickle.dump(top_50_books , open('top_50_books.pkl' , 'wb'))
```

# Collbaritive Recommendation system

***Filter the data to include users who have given more than 100 reviews and books that have more than 50 reviews. We need to keep this filtered data.***

```python
# Filtering users those do more then 100 Ratings
temp = new_df.groupby('User-ID')['Book-Rating'].count() > 100
experienced_user = temp[temp].index
```

```python
# this is dataframe of Experience users
new_df = new_df[new_df['User-ID'].isin(experienced_user)]
```

```python
# Filtering Books those have then 50 Ratings
temp = new_df.groupby('Book-Title')['Book-Rating'].count()  > 50
filter_book = temp[temp].index
```

```python
new_df = new_df[new_df['Book-Title'].isin(filter_book)]
```

```python
print(new_df.shape)
new_df.head()
```

### Create a pivot table with books as the index, users as the columns, and the cell values as the count of book ratings.

```python
pivot_table = new_df.pivot_table(index = 'Book-Title' , columns='User-ID' , values='Book-Rating').fillna(0)
pivot_table
```

```python
pivot_table.info()
```

### Calculating similarity scores (cosine similarity) for the pivot table

```python
# finding cosine similarity for each books with respect to other books
similarity = cosine_similarity(pivot_table)
```

```python
print(similarity.shape)
similarity
```

# Final Recommend Method

```python
def recommend_books(movie_name):
  index = np.where(pivot_table.index == movie_name)[0][0] # fatchine index of given movie

  # creating a empty dataframe to concat each book details and return it
  book_to_recommend = pd.DataFrame( columns = ['Book-Title' , 'Book-Author' , 'Image-URL-M'])

  # fetching top 6 close books of given books
  for movie_index in sorted(list(enumerate(similarity[index])) , reverse = True , key = lambda x : x[1])[1:6]:
    book_name = pivot_table.index[movie_index[0]]  # got book name

    # from book name fatching all the details
    temp = books[books['Book-Title'] == book_name].drop_duplicates(subset='Book-Title')[['Book-Title' , 'Book-Author' , 'Image-URL-M']]

    # concating the book into empty dataframe
    book_to_recommend = pd.concat((book_to_recommend , temp))

  return book_to_recommend
```

```python
recommend_books("24 Hours")
```
